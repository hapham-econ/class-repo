Write a function to find root using the bisection method


```julia
function bisection_method(f, lower_bound, upper_bound, tol)
    difference = 1000
    while difference  >= tol
        f_lower = f(lower_bound)
        f_upper = f(upper_bound)
        println("Upper bound is $upper_bound, lower bound is $lower_bound")
        if sign(f_lower) != sign(f_upper)  # multiply bad, sign good
            mid_point = (lower_bound + upper_bound)/2
            f_mid = f(mid_point)
            if sign(f_lower) != sign(f_mid) 
               upper_bound = mid_point
            elseif sign(f_upper) != sign(f_mid) 
                lower_bound = mid_point
            end
        end
        
        difference = abs(lower_bound - upper_bound)
    end
    println("Close enough! Answer is $upper_bound")
end
```

```julia
 function bisection(f, lower_bound, upper_bound)
     tolerance = 1e-3                                # tolerance for solution
     guess = 0.5*(upper_bound + lower_bound)         # initial guess, bisect the interval
     difference = (upper_bound - lower_bound)/2      # initialize bound difference
     while difference > tolerance                    # loop until convergence
         println("Intermediate guess of $guess.")
         difference = difference/2
         if sign(f(lower_bound)) == sign(f(guess))   # if the guess has the same sign as the lower bound
             lower_bound = guess                     # solution is in the upper half of the interval
             guess = guess + difference
         else                                        # else the solution is in the lower half of the interval
             upper_bound = guess
             guess = guess - difference
         end
     end
     println("The root of f(x) is at $guess.")
 end
```

```julia
function value_iteration(f, x, tol, max_iter)
    error = 1000
    iter = 0
    x_new = 1000
    while error >= tol 
        value = f(x)
        x_new = value
        error = abs(x_new - x)
        x = x_new
        println("new guess is $x_new, error is $error")
        iter += 1
        if iter >= max_iter
            println("too many iterations")
            break
        end
    end
    
    if iter >= max_iter
        println("fail to reach fixed point")
    else
        println("Solved! The answer is $x_new")
    end
    return(x_new)
end
```

```julia
f(x) = 3 + x - x^2
```

```julia
value_iteration(f, 2, 1e-3, 3000)
```

```julia
using ForwardDiff
g(ff,x0) = ForwardDiff.derivative(ff,x0)
function newton(f, x, tol, max_iter)
    error = 1000
    iter = 0
    x_new = 1000
    while error >= tol 
        value = f(x)
        diff = g(f,x)
        x_new = x - (1/diff)*value
        error = abs(x_new - x)
        x = x_new
        println("new guess is $x_new, error is $error")
        iter += 1
        if iter >= max_iter
            println("too many iterations")
            break
        end
    end
    
    if iter >= max_iter
        println("fail to reach fixed point")
    else
        println("Solved! The answer is $x_new")
    end
    return(x_new)
end
```

```julia
f(x) = x^3
newton(f, 2, 1e-5, 3000)
```

```julia
function secant(f, x, tol, max_iter)
    error = 1000
    iter = 0
    x_new = 1000
    x_second = x - 1
    while error >= tol 
        value = f(x)
        value_second = f(x_second)
        x_new = x - ((x-x_second)/(value-value_second))*value
        error = abs(x_new - x)
        x_second = x
        x = x_new
        println("new guess is $x_new, error is $error")
        iter += 1
        if iter >= max_iter
            println("too many iterations")
            break
        end
    end
    
    if iter >= max_iter
        println("fail to reach fixed point")
    else
        println("Solved! The answer is $x_new")
    end
    return(x_new)
end
```
